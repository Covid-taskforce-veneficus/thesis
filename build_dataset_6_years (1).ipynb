```python
df1 = pd.read_csv("dataset_edition(4)(test_version)(nomissingvalues).csv")
df1.shape
```

```python
df1 = df1.iloc[: , 1:]  # run this to delete the irrelevant index columns
```

```python
df1
```

```python
# select the years 2010 till 2014 (more not possible because 2014 + 6 = 2020)


df1_uni = np.unique(df1["Ticker_x"])

data = pd.DataFrame()

for i in df1_uni:
    x = df1[df1["Ticker_x"] == i]
    
    for j in x["year"]:
        y = x[x["year"] == j]
        if j <= 2014:
            data = pd.concat([data, y])
    
data = data.reset_index()
data.iloc[1:20,:]
```

```python
#select target values 2016 till 2020

df1_uni = np.unique(df1["Ticker_x"])

data2 = pd.DataFrame()

for i in df1_uni:
    x = df1[df1["Ticker_x"] == i]
    for j in x["year"]:
        y = x[x["year"] == j]
        if j >= 2016:
            data2 = pd.concat([data2, y])

data2.head()
```

```python
#sort the dataset on alphabet and year

data2 = data2.sort_values(by=['Ticker_x', "year"])
data = data.sort_values(by=['Ticker_x',"year"])
```

```python
#select only the open column, which will be the closing price in our new dataset

closed = data2["Open"]
closed = closed.reset_index()
closed = closed.iloc[0:,1:2]
closed.iloc[1:20,:]
```

```python
data["close"] = closed  # add the close price to the dataset
```

```python
data["6_years"] = round((data["close"] - data["Open"]) / data["Open"],4)  # calculate the difference between open and close stock price
```

```python
data = data.drop(columns = "profit/loss")  # drop the profit/loss since we have made a new proft/loss
```

```python
data = round(data,4)  # round the data
```

```python
data.to_csv("6_year_dataset(1).csv")
```
