### Here I deleted companies who had lots of missing values

```python
import numpy as np
import pandas as pd
import os
import missingno as msno
```

```python
os.getcwd()
os.chdir("C:\\Users\\Jacco Bruin .LAPTOP-ACFSBHIL\\THESIS")
df1 = pd.read_csv('dataset_8_11_21.csv')
df1 = df1.replace(0, np.nan)   #change zeros to NAN
msno.bar(df1, figsize=(12, 6), fontsize=12, color='steelblue')   #check how much data is missing
```

```python
# Check if there are companies who have 0 information for certain features

df1_uni = np.unique(df1["Ticker_x"])

data = pd.DataFrame()

for i in df1_uni:
    x = df1[df1["Ticker_x"] == i]
    y = x.sum()
    for j in y:
        if j == 0:
            data = pd.concat([data, x])
            break
```

```python
data_uni = np.unique(data["Ticker_x"])

print(len(df1_uni))         #total companies
print(len(data_uni))        #companies with 0 information for certain features (122)
data_uni
```

```python
# delete those companies 

df1_1 = pd.DataFrame()

for i in df1_uni:
    x = df1[df1["Ticker_x"] == i]
    if i not in data_uni:
        df1_1 = pd.concat([df1_1, x])
```

```python
df1_1.to_csv("dataset_edition(1).csv")
```

```python
df1_1 = df1_1.replace(0, np.nan)

print(df1.isna().sum().sum())
print(df1_1.isna().sum().sum())
msno.bar(df1_1, figsize=(12, 6), fontsize=12, color='steelblue')

#missing values went from 9963 to 5297; however, still missing values
```
