```python
import os
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score
from sklearn import preprocessing
from sklearn.model_selection import RandomizedSearchCV
```

```python
os.getcwd()
os.chdir("C:\\Users\\Jacco Bruin .LAPTOP-ACFSBHIL\\THESIS")
df1 = pd.read_csv('1_year_dataset(3)(without_outliers).csv') 
#df2 = pd.read_csv('6_year_dataset(2)(with_cat).csv') 

df12 = df1["Ticker_x"]

df1 = df1.set_index("Ticker_x")

df1.head()
```

```python
#scale data and add columns and index to it 

df11 = df1.columns
x = df1.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df1 = pd.DataFrame(x_scaled)
df1.columns = [df11]
df1.index = [df12]
```

```python
X = df1.iloc[:,1:19]   
y = df1.iloc[:,-2]
X.columns
```

```python
# select the mean of the target values and make this equal to rows of dataset

x = df1.shape[0]
y_pred = [y.mean()] * x   
```

```python
# evalute the score

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

MSE = mean_squared_error(y,y_pred)
MAE = mean_absolute_error(y,y_pred)
R2 =  r2_score(y,y_pred)  

n = len(y)
p = X.shape[1]

Adj_R2 = 1-(1-R2)*(n-1)/(n-p-1)


print(MSE)
print(MAE)
print("R2_adj of test Model with Cross Validation is:", Adj_R2)
```
