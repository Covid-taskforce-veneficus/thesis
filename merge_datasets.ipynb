## Not all companies have the same amount of data. Some companies have data from 2008 till 2020, and some have data from 2015 till 2020. To select the most complete data only companies who had data from 2010 till 2020 were selected. 

```python
#select only the year 2010 till 2020. 


df1_uni = np.unique(df1["Ticker"])

data = pd.DataFrame()

for i in df1_uni:
    x = df1[df1["Ticker"] == i]
    min_value = x["Fiscal Year"].min()
    #print(min_value)
    if min_value == 2008:    # i also want to select companies who have data from 2008 till 2020, therefore this line of code 
        min_value += 2
        x = x.drop(x[x["Fiscal Year"] == 2008].index)
    if min_value == 2009:    # i also want to select companies who have data from 2009 till 2020, therefore this line of code 
        min_value += 1
        x = x.drop(x[x["Fiscal Year"] == 2009].index)
    if (min_value == 2010) and (x["Fiscal Year"].max() == 2020) and (len(x["Fiscal Year"]) == 11):
        data = pd.concat([data, x])
     
    
print(len(data))      
pd.DataFrame(data)   
```

```python
data_uni = np.unique(data["Ticker"])  
len(data_uni)                           #this gave 827 companies 
```

```python
df8 = pd.read_csv("dataset_only_october.csv")
df8_1 = df8[["Ticker","Open", "year"]]
```

```python
# append the openingsprice to the financial ratio + macroeconomic variable model. So only years 2010 till 2020 are selected 

df7_uni = np.unique(df7["Ticker"])

data = pd.DataFrame()

for i in df7_uni:
    x = df7[df7["Ticker"] == i]
    y = df8_1[df8_1["Ticker"] == i]
    total_data = x.merge(y, on = "year", how = "left")
    data = pd.concat([data, total_data])

data.to_csv("tete.csv")  # save the dataset, so result till now are saved
```

```python
df10 = pd.read_csv('tete.csv')  #open the dataset again
```

```python
# add the percentage difference between 2010 - 2011, 2010 - 2012, 2010 = 2013 etc

df10_uni = np.unique(df10["Ticker_x"])
data = pd.DataFrame()

for i in df10_uni:
    
    x = df10[df10["Ticker_x"] == i]
    price_2010 = x["Open"].iloc[0]
    total_result = []
    for i in x["Open"]:
        result = round((i - price_2010) / price_2010,4)
        total_result.append(result)
    x["profit/loss"] =  total_result
    data = pd.concat([data, x])

data.to_csv("TEST_7_11_21.csv")    # save the dataset, so result till now are saved
```
