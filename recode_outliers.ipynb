### In this file we deal with the outliers per dataset 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats.mstats import winsorize
```

```python
os.getcwd()
os.chdir("C:\\Users\\Jacco Bruin .LAPTOP-ACFSBHIL\\THESIS")
data = pd.read_csv('6_year_dataset(1).csv')             # change here the dataset: 6_year_dataset, 3_year_dataset, or 1_year_dataset
data = data.set_index("Ticker_x")
data = data.iloc[:,1:]
data = data.iloc[:,1:]   # if there is index column delete this one first 
data.head()
```

```python
#plots all the features that have a high skewnesss

outliers = []
for i in data.columns:
    if skew(data[i]) > 3 or skew(data[i] < -3):
        plt.hist(data[i], bins = 20)
        plt.title(i)
        plt.show()
        print(skew(data[i]))
        outliers.append(i)
```

```python
for i in outliers:
    data[i] = winsorize(data[i], limits=(0.05, 0.05))
```

```python
for i in data.columns:
    plt.hist(data[i], bins = 20)
    plt.title(i)
    plt.show()
    print(skew(data[i]))
```

```python
data.to_csv("6_year_dataset(4)(without_outliers).csv")   # change this per dataset analysed 
```
